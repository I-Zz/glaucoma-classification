```python
import os
import pandas as pd

# Paths to your base image folder and CSV
image_base_folder = '/kaggle/input/jraigs-dataset/justRAIGS'  # Adjust this to your base folder where subfolders (0, 1, etc.) are
csv_file = '/kaggle/input/jraigs-dataset/justRAIGS/JustRAIGS_Train_labels.csv'  # Your original CSV file

# Load the CSV file
df = pd.read_csv(csv_file , sep = ";" , encoding ="utf-8")

# Create a dictionary to store image paths with file names as keys (without extensions)
image_paths = {}

# Function to find images, considering the possibility of an extra folder layer, and ignoring .db files
def collect_image_paths(base_folder):
    valid_extensions = ('.png', '.jpg', '.jpeg', '.bmp', '.tiff')  # Add any other image formats you expect
    for root, dirs, files in os.walk(base_folder):
        for file in files:
            # Only process files with valid image extensions, making extension check case-insensitive
            if file.lower().endswith(valid_extensions):
                # Get the file name without extension (preserve original case)
                eye_id = os.path.splitext(file)[0]

                # Store the full path in the dictionary, case-sensitive
                image_paths[eye_id] = os.path.join(root, file)

                
                
# Call the function to collect paths from all subfolders
collect_image_paths(image_base_folder)

# Define a function to map Eye ID to the image path
def assign_image_path(eye_id):
    return image_paths.get(eye_id, None)  # Return the image path if it exists, else None

# Apply the function to assign image paths to the Eye ID
df['Image Path'] = df['Eye ID'].apply(assign_image_path)

# Save the new CSV with image paths
df.to_csv('/kaggle/working/JustRAIGS_Train_labels_with_paths.csv', index=False)

print("CSV updated with image paths!")
```

```python
folder_path = '/kaggle/input/jraigs-dataset/justRAIGS'  # Adjust this to your folder path

# Function to count files in a folder (including subfolders)
def count_files_in_folder(folder_path):
    file_count = 0
    for root, dirs, files in os.walk(folder_path):
        file_count += len(files)
    return file_count

# Get the file count
total_files = count_files_in_folder(folder_path)

print(f"Total number of files: {total_files}")
```
